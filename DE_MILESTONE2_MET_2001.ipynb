```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import math

#Web Source (downloaded dataframe) -> 
#https://public.opendatasoft.com/explore/dataset/georef-united-kingdom-local-authority-district/table/?disjunctive.ctry_code&disjunctive.ctry_name&disjunctive.rgn_code&disjunctive.rgn_name&disjunctive.ctyua_code&disjunctive.ctyua_name&disjunctive.lad_code&disjunctive.lad_name

dfWeb = pd.read_excel("georef-united-kingdom-local-authority-district.xls", header=0)
dfWeb
```

```python
#Creating Dataframe with desired features from the dataframe obtained from the web 
df2=pd.DataFrame()
df2['local_authority_district']=dfWeb['Official Name Local authority district']
df2['local_authority_district_region']=dfWeb['Official Name Region']
df2['local_authority_district_country']=dfWeb['Official Name Country']
df2
```

```python
#--------------------------------------------------------------------------------------------------------------------------------------------------------
```

```python
#Merging Dataframes (to add new features: local_authority_district_region & local_authority_district_country)
df = pd.read_csv("new_df.csv")
df= pd.merge(df,df2)
df
```

```python
#No scaling needed as features added are categorical
#Features will not be encoded to produce clear & understandable dataframes to the human eye 
```

```python
#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Questions & Visualizations
```

```python
def plotBarGraph(data, col, rot, title):
    sns.countplot(data=df, x =col)
    plt.xticks(
    rotation=rot, 
    horizontalalignment='right',
    fontsize='medium'  
        )
    plt.title(title)
    plt.show()
```

```python
#Question 1: what is the accident frequency per region in the UK?

#Answer 1:
plotBarGraph(df, 'local_authority_district_region', 40, 'Bar Graph for Accident Frequency Per Region')  
```

```python
#Question 2: what is the accident frequency per country in the UK?

#Answer 2:
plotBarGraph(df, 'local_authority_district_country', 40, 'Bar Graph for Accident Frequency Per Country')  
```
