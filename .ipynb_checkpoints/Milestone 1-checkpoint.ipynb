# 3 - Cleaning Data

## Observing Custom Missing Values & Cleaning it (change to null)

```python
#Observing Values & Repetition of Every Column
def observeValues(df):
    cols= df.columns.tolist()
   
    for x in cols:
        y=df.groupby([x]).size()
        print(y)
        print()
        

def observeNumericalValues(df):
    cols= df.columns.tolist()
   
    for x in cols:
        if (df[x].dtype=="int64" or df[x].dtype=="float64" or df[x].dtype=="int32"):
            y=df.groupby([x]).size()
            print(y)
            print()

def observeCategoricalValues(df):
    cols= df.columns.tolist()
   
    for x in cols:
        if (df[x].dtype=="object"):
            y=df.groupby([x]).size()
            print(y)
            print()  

observeValues(df)
```

```python
df.info()
```

```python
#Cleaning Custom Missing Values
#As you can see "Data Missing" & -1 & '-1' represents missing data -> change to null
#first_road_class is C or Unclassified. These roads do not have official numbers so recorded as zero -> change to 0

def cleanArbitrary(df):
    df.replace(-1, np.nan, inplace=True)
    df.replace('-1', np.nan, inplace=True)
    df.replace('Data missing or out of range', np.nan, inplace=True)
    df.replace('first_road_class is C or Unclassified. These roads do not have official numbers so recorded as zero ', 0, inplace=True)
 

cleanArbitrary(df)
```

## Observing Missing and duplicate Data

```python
#Observing Missing Data (percentages)

def getMissingPercentages(df):
    
    perc_null = df.isnull().sum() / len(df) * 100
    
    def getColsMissingLowPercentage(perc_null):
        colsMissingLowPercentage=[]
        print("Low Missing Data Percentages")
        for index,value in perc_null.items():
            if (value>0 and value<=5):
                print(str(round(value,2)) + " %" + " | " +  str(index)   )
                colsMissingLowPercentage.append(index)
        print()        
        return colsMissingLowPercentage

    def getColsMissingHighPercentage(perc_null):
        colsMissingHighPercentage=[]
        print("High Missing Data Percentages")
        for index,value in perc_null.items():
            if (value>5):
                print(str(round(value,2)) + " %" + " | " +  str(index)   )
                colsMissingHighPercentage.append(index)
        print()
        return colsMissingHighPercentage
       
    low=getColsMissingLowPercentage(perc_null)
    high= getColsMissingHighPercentage(perc_null)
    # return perc_null,low,high


getMissingPercentages(df)
```

## Handling Missing data

```python
#Cleaning Missing Data 1

#Column(s): location_easting_osgr,location_northing_osgr,longitude, latitude
#Reason: MCAR 
#Percentage Missing: 0.35%
#Action: drop, when these 4 columns are missing (using how=all), to make sure they're are the same accidents that are missing the locations
#Action Reason: low missing ratio & any imputation would result in false info & you can't get from other columns

df.dropna(axis='index', how='all', subset=['location_easting_osgr', 'location_northing_osgr', 'longitude', 'latitude'], inplace=True)


getMissingPercentages(df)
```

```python
#Cleaning Missing Data 2

#Column(s): (in this cell)
#Reason: MCAR 
#Percentage Missing: all <0.25% except one 1.42%
#Action: drop 
#Action Reason: low missing ratio & any imputation would result in false info & you can't get from other columns


df.dropna(axis='index', subset=['pedestrian_crossing_human_control'], inplace=True)
df.dropna(axis='index', subset=['junction_detail'], inplace=True)
df.dropna(axis='index', subset=['pedestrian_crossing_human_control'], inplace=True)
df.dropna(axis='index', subset=['pedestrian_crossing_physical_facilities'], inplace=True)
df.dropna(axis='index', subset=['weather_conditions'], inplace=True)
df.dropna(axis='index', subset=['road_surface_conditions'], inplace=True)
df.dropna(axis='index', subset=['special_conditions_at_site'], inplace=True)
df.dropna(axis='index', subset=['carriageway_hazards'], inplace=True)

df=df.reset_index(drop=True)

getMissingPercentages(df)
```

```python
#Cleaning Missing Data 3

#Column(s): (in this cell)
#Reason: MAR 
#Percentage Missing: <3% & 10%(1)
#Action: multivariate mode imputation
#Action Reason: we can infer these variable from other variables

def multivariateModeImputation(df,independentVar,changingVar):
    modeList=df.groupby([independentVar])[changingVar].agg(pd.Series.mode) # get you series with index values (unique independent variable value) & 
    # the changing variable mode correspondonding to each unique independent variable value
    
    df_subset= df[[changingVar,independentVar]]

    for i in range(len(df_subset[changingVar])):
        currentChangingVarValue= df_subset[changingVar][i]
      
        if (currentChangingVarValue is np.nan):
            currentIndependentVarValue= df_subset[independentVar][i]
            
            currentModeValue=modeList[currentIndependentVarValue]
            if not(type(currentModeValue) is np.ndarray): #currentModeValue will be [] (type: np.ndarray), if there's no mode
                df_subset[changingVar][i]= currentModeValue
                
            else:#no mode available  -> insert abitrary value (-1) indicating missing (if np.ndarray got inserted it will mess up dataframe
                df_subset[changingVar][i]=-1
              


    df[changingVar]= df_subset[changingVar]
    return df

    
multivariateModeImputation(df,'local_authority_district', 'first_road_number')
multivariateModeImputation(df,'local_authority_district', 'local_authority_highway')
multivariateModeImputation(df,'first_road_class', 'road_type')
multivariateModeImputation(df,'accident_severity', 'did_police_officer_attend_scene_of_accident')
multivariateModeImputation(df,'first_road_class', 'trunk_road_flag')



observeValues(df)
getMissingPercentages(df)
```

```python
#Cleaning Missing Data 4

#Column(s): (in this cell)
#Reason: MAR (there's a relation between other variable like junction_detail (Not at junction or within 20 metres = no junction control or 2nd road no or class)
#Percentage 39% (3)
#Action: fill with arbitrary value (-1) inidicates missing  
#Action Reason: high missing ratio 

df.junction_control=df.junction_control.fillna('-1')
df.second_road_number=df.second_road_number.fillna(-1)
df.second_road_class=df.second_road_class.fillna('-1')


#Cleaning Missing Data 5

#Column(s): (in this cell)
#Reason: MCAR 
#Percentage Missing: 10%
#Action: fill with arbitrary value (-1) inidicates missing
#Action Reason: high missing ratio

df.lsoa_of_accident_location=df.lsoa_of_accident_location.fillna('-1')
getMissingPercentages(df)
```

## Observing outliers

```python
#Outlier Detection Done for numerical columns only (if applicable)
observeNumericalValues(df)
```

```python
#As you can see, accident_year has only 1 unique value
#Outlier detection will done for -> number_of_vehicles, number_of_casualties, speed_limit

def detectOutlierIQR(df, variable):
    sns.boxplot(df[variable])
    plt.show()
    
    Q1 = df[variable].quantile(0.25)
    Q3 = df[variable].quantile(0.75)
    IQR = Q3 - Q1
    print(IQR)
    cut_off = IQR * 1.5
    lower = Q1 - cut_off
    upper =  Q3 + cut_off
    print(lower,upper)

    df1 = df[df[variable]> upper]
    df2 = df[df[variable] < lower]
    sum1= df1.shape[0]+ df2.shape[0]
    perc= sum1/ len(df)*100
    print('Total number of outliers are', sum1)
    print('Outliers Percentage', perc)
    print()
    
detectOutlierIQR(df,"number_of_vehicles")
detectOutlierIQR(df,"number_of_casualties")
detectOutlierIQR(df,"speed_limit")
```

```python
#Outliers Observations
#As you can see 'speed_limit' has no outliers so nothing will be done
#no_of_casualties has high outlier rate, so we'll check the outliers of 'number_of_vehicles' & 'number_of_casualties' together using local outlier factor method
```

```python
from sklearn.neighbors import LocalOutlierFactor

def detectOutlierLOF(df, firstVar, secondVar):
    clf = LocalOutlierFactor()
    X = df[[firstVar,secondVar]].values
    y_pred = clf.fit_predict(X)
    plt.figure(figsize=(12,12))
    # plot the level sets of the decision function

    in_mask = [True if l == 1 else False for l in y_pred]
    out_mask = [True if l == -1 else False for l in y_pred]

    plt.title("Local Outlier Factor (LOF)")
    # inliers
    a = plt.scatter(X[in_mask, 0], X[in_mask, 1], c = 'blue',edgecolor = 'k', s = 30)
    # outliers
    b = plt.scatter(X[out_mask, 0], X[out_mask, 1], c = 'red',edgecolor = 'k', s = 30)

    plt.axis('tight')
    plt.xlabel(firstVar);
    plt.ylabel(secondVar);
    plt.show()
    
    return X,in_mask,out_mask

def removeOutliersLOF(df,firstVar, secondVar, X,in_mask,out_mask):

    noOfOutliers= len(X[out_mask,0])
    totalVals= len(df)
    outliersPerc= noOfOutliers/totalVals * 100
    
    print("Number of Outliers:", noOfOutliers)
    print("Outliers Percentage:", round(outliersPerc,2), "%")
    print()

    list1=X[out_mask,0].copy() #1st variable values (outliers)
    list2= X[out_mask,1].copy() #2nd variable values (outliers)
    list3=[] #make list of pairs (1st var value, 2nd var value)
    #each combination is an outlier -> ex: value1=X[out_mask,0][0] , value2= X[out_mask,1][0]     
    #if value1 & 2 are present in the same row then this row has an outlier (these 2 variables)

    for i in range(len(list1)):
        obj=(list1[i],list2[i]) 
        list3.append(obj)

    list3 = [*set(list3)] # remove duplicate pairs  (to get unique outlier combinations)
  
    #df2=df.copy()
    print("Before Handling")
    print(df.shape)
    #print(df2.shape)

    for i in range(len(list3)):
        firstVal,secondVal= list3[i] # current outlier pair
        p=df[(df[firstVar]==firstVal) & (df[secondVar]==secondVal)]
        df=pd.concat([df, p, p]).drop_duplicates(keep=False)  #concatinate & remove duplicates (both) -> remove rows of dataframe from original datafram

    df=df.reset_index(drop=True)    
    print("After Handling")
    print(df.shape)
    return df
    
```

```python
x,y,z = detectOutlierLOF(df,'number_of_vehicles','number_of_casualties')
```

```python
#Sepecial Ordinal Encoding (new numerical value has relation with old values)


newVals = {"Yes" : 1,'No' : 0}
df["did_police_officer_attend_scene_of_accident"]=df["did_police_officer_attend_scene_of_accident"].map(newVals)

newVals = {"Trunk (Roads managed by Highways England)" : 1,'Non-trunk' : 0}
df["trunk_road_flag"]=df["trunk_road_flag"].map(newVals)

newVals = {"Fatal" : 2,'Serious' : 1, 'Slight': 0}
df["accident_severity"]=df["accident_severity"].map(newVals)

```

```python
#Label Encode
from sklearn import preprocessing

def labelEncode(df, var):
    le= preprocessing.LabelEncoder()
    
    le.fit(df[var])
    mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    df[var] = le.fit_transform(df[var])
  
    return mapping


colsToLabelEncode= ['first_road_class',  'road_type',  'junction_detail', 'junction_control', 'second_road_class',  'pedestrian_crossing_human_control', 'pedestrian_crossing_physical_facilities', 'light_conditions', 'weather_conditions', 'road_surface_conditions', 'special_conditions_at_site', 'carriageway_hazards', 'urban_or_rural_area']
mappings=[]



for col in colsToLabelEncode:
    if (df.dtypes[col] == np.object):
       
        map1 = labelEncode(df,col)
        obj={"Column Name": col, "Map": map1}
        mappings.append(obj)
       
    
# print(mappings)
```

## 4.4 - Adding more columns

```python
#Hour & Day-Part Column (ADD)

timeList = df['time'].copy()
hourList=[]
dayPartList=[]

for time in timeList:
    timeSep= time.split(":")
    hour= int(timeSep[0])
   
    
    dayPart=''
    if (hour>=22 or hour<=4):
        dayPart='Night'
    if (hour>=5 and hour<=12):
        dayPart='Morning'    
    if (hour>=13 and hour<=16):
        dayPart='Afternoon'
    if (hour>=17 and hour<=21):
        dayPart='Evening'
    
    hourList.append(hour)
    dayPartList.append(dayPart)


df['hour']= hourList
df['day_part']= dayPartList
```
